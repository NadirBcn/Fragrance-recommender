```python
import pandas as pd
from bs4 import BeautifulSoup
import csv
```

```python
#reading in the urls gathered in 01_getting_soups.ipynb
urls = pd.read_csv('C:/Users/Nadir/Desktop/B/Projects/Fragrance_project/data/product_urls.csv')
urls
```

```python
#opening a new csv file with headers
header = ['name', 'brand', 'category', 'price', 'ingredients', 'no_reviews', 'hearts', 'size1', 'size2', 'url']

with open('C:/Users/Nadir/Desktop/B/Projects/Fragrance_project/data/product_info.csv', "w", newline='') as f:
    writer = csv.writer(f, delimiter=',')
    writer.writerow(header) # write the header
```

```python
for i in range(0,1440):
# for i in urls.index:
    products = [] #empty list to append the dictionary in to before passing in to a DataFrame
    #setting the path to the the html files
    path = "C:/Users/Nadir/Desktop/B/Projects/Fragrance_project/data/soups/soup"+str(i)+".html"
    #opening and reading the file
    file_path = open(path, 'rb')
    file_read = file_path.read()
    #creating a new soup from it
    soup = BeautifulSoup(file_read)
    #gathering the data from the pages
    try:   
              
        product = {}
        product['name'] = soup.find('span', {'class': 'css-1pgnl76 e65zztl0'}).text
        product['brand'] = soup.find('span', {'class': 'css-bpsjlq.eanm77i0'}).text
        product['category'] = urls.category[i]
        product['price'] = soup.find('div', {'class': 'css-1oz9qb'}).text
        product['ingredients'] = soup.find('div', {'class': 'css-1ue8dmw.eanm77i0'})
        #product['ingredients'] = str(ingredients).split('<br/>') #will need to break during cleaning
        product['no_reviews'] = soup.find_all('span', {'class': 'css-1j53ife'})[0].text
        product['hearts'] = soup.find('span', {'class': 'css-jk94q9'}).text
        #the product size has two different formats on the pages
        product['size1'] = size = soup.find('span', {'class': 'css-1hdop86'}).text
        try:
            product['size2'] = soup.find('span', {'class': 'css-7b7t20'}).text
        except:
            product['size2'] = '0'
        product['url'] = urls.URL[i]
    
        #append the empty list to later make in to a dataframe
        products.append(product)
        product_df = pd.DataFrame(products) 
        #append in the csv file that created above
        product_df.to_csv('C:/Users/Nadir/Desktop/B/Projects/Fragrance_project/data/product_info.csv', mode='a', index = False, header = False)

    except: #if the page no longer existed when gathered html
        pass
   
```
