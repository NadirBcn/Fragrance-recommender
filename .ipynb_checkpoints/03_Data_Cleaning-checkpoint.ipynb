```python
# Adding a new Family column

df['Family'] = ""

for row in df.index:
    fam = re.search('Family: (.*)Scent',df['Description'][row])
    if fam is not None:
        df['Family'][row] = fam.group(1)
    else:
        df['Family'][row] = 'Unknown'
```

```python
# Adding a new scent column

df['Scent'] = ""

for row in df.index:
    scen = re.search('Type: (.*)Key',df['Description'][row])
    if scen is not None:
        df['Scent'][row] = scen.group(1)
    else:
        df['Scent'][row] = 'Unknown'
```

```python
# Adding a new key notes column

df['Notes'] = ""

for row in df.index:
    notes = re.search('Notes: (.*)Fragrance Description:',df['Description'][row])
    if notes is not None:
        df['Notes'][row] = notes.group(1)
    else:
        df['Notes'][row] = 'Unknown'
```

```python
C = df['overall_rating'].mean()
C
```

```python
# Minimum review amount (Top 10%)

m = df['max_review'].quantile(0.9)
m
```

```python
# Assigning the fragrance with the most reviews (Top 10%) to a list

list_fragrance = df.copy().loc[df['max_review'] >= m]

list_fragrance.shape
```

```python
# Function that calculate the "score" of each perfume 

def weighted_ratings(df,m=m,C=C):
    v = df['max_review']
    R = df['overall_rating']
    # Imdb forumla 
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
list_fragrance['Score'] = list_fragrance.apply(weighted_ratings,axis=1)
```

```python
list_fragrance.head(3)
```

```python
# Re arrange the new <Score> column

cols = list_fragrance.columns.to_list()
cols2 = df.columns.to_list()

cols = cols[:7] + cols[-4:] + cols[7:-4]
cols2 = cols2[:7] + cols2[-4:] + cols2[7:-4]

print(cols2)

list_fragrance = list_fragrance[cols]
df = df[cols2]
```
